# Data

```python
data = pd.read_table('data0.tsv')
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['gold'] = data['gold'].str.replace(',','.').astype(np.float64)
data['silver'] = data['silver'].str.replace(',','.').astype(np.float64)
data['platinum'] = data['platinum'].str.replace(',','.').astype(np.float64)
data['palladium'] = data['palladium'].str.replace(',','.').astype(np.float64)
```

```python
data['silver50'] = data['silver']*50
```

```python
data['date'] = data['date'].str.replace('.', '/')
```

```python
data['date'] = pd.to_datetime(data['date'], format='%d/%m/%y')
```

```python
data.head()
```

```python
data['date'] = data['date'].astype(str)
```

```python
data['date'] = data['date'].str.replace('-', '')
```
